```python
import torch
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
from google.colab import files
from IPython.display import display


uploaded = files.upload()
image_path = next(iter(uploaded))

# Load a smaller BLIP model (for faster inference)
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")


image = Image.open(image_path).convert("RGB")


image = image.resize((256, 256))


display(image)

inputs = processor(image, return_tensors="pt")

# Generate caption with optimized parameters
with torch.no_grad():
    caption_ids = model.generate(**inputs, max_length=30, num_beams=3)  # Reduced parameters
    caption = processor.batch_decode(caption_ids, skip_special_tokens=True)[0]


print("\nðŸ”¹ Generated Caption:", caption)
```
